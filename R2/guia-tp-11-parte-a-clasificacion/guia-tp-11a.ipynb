Las aves par ́asitas de cr ́ıa ponen huevos en nidos de otras especies (hospedador), las
cuales incuban los huevos y cr ́ıan al pich ́on par ́asito.

En un bosque de talas de la provincia
de Buenos Aires hay dos especies hospederas que son indistinguibles a simple vista.

Una
de las principales diferencias entre estas especies radica en el grado de discriminaci ́on y
remoci ́on de huevos par ́asitos de sus nidos.

Una de las especies es **“aceptadora“ de huevos
par ́asitos (Y = 0)**, ya que remueve del nido s ́olo el **30%** de los huevos par ́asitos, mientras
que la otra especie es **“rechazadora“ (Y = 1)** ya que **remueve el 80%** de los huevos par ́asitos
presentes en su nido.

Adem ́as, se sabe que **el 90% de los nidos del bosque corresponden
a la especie “aceptadora“** , mientras que apenas el **10% restante son nidos de la especie
“rechazadora“**.


### Cálculo de probabilidades:

Se elige al azar un nido del bosque y se colocan n = 8
huevos par ́asitos.

Denotemos con X a la variable aleatoria que indica el n ́umero de huevos
removidos del nido.

Asuma que, en cada nido, la remoci ́on (o no) de los diferentes huevos
se realiza de manera independiente.

## 1.
1. Calcule la probabilidad de que 5 de los huevos par ́asitos sean removidos del nido si se
sabe que el nido parasitado es de la especie aceptadora.

```r
# Es puede ser modelado con una binomial de parametros (n,p)

n = 8
p.A = 0.9
p.R = 0.1
p.remu.A = 0.3
p.remu.R = 0.8
```

```r
k = 5
p.remu.5.A = choose(n, k) * p.remu.A^k * (1-p.remu.A)^(n-k)
p.remu.5.A
```

2. Calcule la probabilidad de que x de los huevos ajenos sean removidos del nido si se
sabe que el nido parasitado es de la especie aceptadora, para 0 ≤ x ≤ 8, y represente
las probabilidades calculadas en un diagrama de barras.
  
  Incluya un t ́ıtulo pertinente en el gr ́afico.

```r
p.remu.x.A = rep(NA, n+1)
for(k in 0:n){
    p.remu.x.A[k+1] = choose(n, k) * p.remu.A^k * (1-p.remu.A)^(n-k)
}
names(p.remu.x.A) <- 0:n
```

```r
barplot(p.remu.x.A, main="Probabilidad de remover X huevos por especie Aceptadora",
        xlab="Huevos removidos", ylab="Probabilidad", col="steelblue")
```

3. Calcule la probabilidad de que 5 de los huevos par ́asitos sean removidos del nido si se
sabe que el nido parasitado es de la especie rechazadora.

```r
k = 5
p.remu.5.R = choose(n, k) * p.remu.A^k * (1-p.remu.A)^(n-k)
p.remu.5.R
```

4. Calcule la probabilidad de que x de los huevos par ́asitos sean removidos del nido si se
sabe que el nido parasitado es de la especie rechazadora, para 0 ≤ x ≤ 8, y representa
las probabilidades calculadas en un diagrama de barras. Incluya un t ́ıtulo pertinente
en el gr ́afico.

```r
p.remu.x.R = rep(NA, n+1)
for(k in 0:n){
    p.remu.x.R[k+1] = choose(n, k) * p.remu.R^k * (1-p.remu.R)^(n-k)
}
names(p.remu.x.R) <- 0:n
```

```r
barplot(p.remu.x.R, main="Probabilidad de remover X huevos por especie Rechazadora",
        xlab="Huevos removidos", ylab="Probabilidad", col="red")
```

5. Calcule la probabilidad de que 5 de los huevos par ́asitos sean removidos del nido.

```r
p.remu.5 = p.remu.5.A * p.A + p.remu.5.R * p.R
p.remu.5
```

6. Calcule la probabilidad de que x de los huevos par ́asitos sean removidos del nido,
para 0 ≤ x ≤ 8, y represente las probabilidades calculadas en un diagrama de barras.
Incluya un t ́ıtulo pertinente en el gr ́afico.

```r
p.remu.x = rep(NA, n+1)
for(k in 0:n){
    p.remu.x.A = choose(n, k) * p.remu.A^k * (1-p.remu.A)^(n-k)
    p.remu.x.R = choose(n, k) * p.remu.R^k * (1-p.remu.R)^(n-k)
    p.remu.x[k+1] = p.remu.x.A * p.A + p.remu.x.R * p.R
}
names(p.remu.x) <- 0:n
```

```r
barplot(p.remu.x, main="Probabilidad de remover X huevos",
        xlab="Huevos removidos", ylab="Probabilidad", col="violet")
```

7. Haga la tabla de la funci ́on de probabilidad conjunta del vector (X, Y ).

```r
#X = Huevos removidos
#Y = Especie: 0=Aceptadora, 1=Rechazadora
tabla = data.frame(matrix(NA, nrow = 2, ncol = n+1))
colnames(tabla) <- c('X0','X1','X2','X3','X4','X5','X6','X7','X8')
rownames(tabla) <- c('Y0.A','Y1.R')
tabla
```

```r
p.remu.x = rep(NA, n+1)

for(k in 0:n){
    p.remu.x.A = choose(n, k) * p.remu.A^k * (1-p.remu.A)^(n-k)
    p.remu.x.R = choose(n, k) * p.remu.R^k * (1-p.remu.R)^(n-k)
    tabla['Y0.A', k+1] = p.remu.x.A
    tabla['Y1.R', k+1] = p.remu.x.R
}

tabla
```

```r
barplot(as.numeric(tabla['Y0.A',]), main="Probabilidad de remover X huevos",
        xlab="Huevos removidos", ylab="Probabilidad", col=rgb(0,0,1,0.5))

barplot(as.numeric(tabla['Y1.R',]), main="Probabilidad de remover X huevos",
        xlab="Huevos removidos", ylab="Probabilidad", col=rgb(1,0,0,0.5), add=T)
```
