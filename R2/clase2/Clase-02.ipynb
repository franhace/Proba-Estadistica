Implementar una funci ́on funcion.h que dado un numero devuelva el valor de la funcion h en ese numero.

```r
# Notar que es la ditribucion acumulada de una distribucion uniforme
funcion.h <- function(x, a=10, b=15){
    if(x < a){
        h <- 0
    }else if(x >= a & x <= b){
        h <- (x - a) / (b - a)
    }
    else{
        h <- 1
    }
    return(h)
}
```

```r
suma.cubilete <- function(dados=5){
    suma <- sum(sample(x=1:6, size=dados, replace=TRUE))
    return(suma)
}
```

```r
datos <- lapply(rep(5, 1000), FUN=suma.cubilete)
```
