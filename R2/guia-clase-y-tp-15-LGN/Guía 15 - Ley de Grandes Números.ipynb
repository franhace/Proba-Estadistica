b) Consideramos n = 2 variables aleatorias X 1 y X 2 independientes con distribución
N (5, 4), la suma y el promedio de ambas, es decir,

$$S_2 = X_1 + X_2 \ \ \ \ \text{y} \ \ \ \ \bar X_2 = \frac{X_1+X_2}{2}$$

Generamos n = 2 datos (independientes) correspondientes a una variable aleatoria con distribución N (5, 4) y computamos suma y promedio.

Replicamos N rep = 1000 veces y realizamos los histogramas de S n y X n a partir de los N rep = 1000

```r
n <- 2
Nrep <- 10000
mu <- 5
var <- 4
sd <- sqrt(var)
```

c) Repetir comenzando con otra semilla y superponer el nuevo gráfico utilizando un color diferente.

```r
Nrep <- 2000
```

```r
library(ggplot2)
```

```r
set.seed(42)
muestra.unif <- runif(n=Nrep, min=0, max=2)

cummean <- rep(NA, Nrep)
for(i in 1:Nrep){
    cummean[i] <- mean(muestra.unif[1:i])
}

datos <- data.frame(means=cummean)

ggplot(datos, aes(x=1:Nrep, y=means, colour=abs(means-1))) +
        geom_point(size=0.5)+
        geom_hline(yintercept=1, linetype="dashed", color = "red", size=0.5)+
        scale_color_gradient(low="blue", high="red")+
        ggtitle("Promedios acumulados de muestra Uniforme (0,2)") +
        xlab("Primeros n elementos") + ylab("Promedio")
```
