## 1.

$$(b-a) \frac{\sum_{i=1}^{n} \frac{1}{\sqrt{2\pi}} e^{-\frac{u_i^2}{2}}}{n} $$

```r
numeric_approx_1 <- function(n, a,b){
    unifs <- runif(n, min=a, max=b)
    return ((b-a)/n * sum(1/sqrt(2*pi) * exp(-0.5*unifs^2)))
}
```

### i.

```r
a <- -1.96
b <-  1.96
numeric_approx_1(a,b,n=100)
numeric_approx_1(a,b,n=1000)
numeric_approx_1(a,b,n=50000)
```

```r
pnorm(q=b)-pnorm(q=a)
```

```r
gridx <- 1:5000
plot(gridx,lapply(gridx, numeric_approx_1, a, b), pch=20, col='steelblue')
abline(h=pnorm(q=b)-pnorm(q=a), col='red')
```

### b) 
$Z_1, Z_2, \dots, Z_n \ \ i.i.d.$

$Z_i \sim \mathcal N(0,1) $

$$\frac 1 n \sum_{i=1}^{n} \mathbb 1_{(a,b)} (Z_i)$$

> Es razonable ya que se cuenta la proporci√≥n de valores que caen en el intervalo (a,b), donde estos valores son generador a partir de una distribuci√≥n Normal(0,1), o sea, se distribuyen siguiendo esa funci√≥n de probabilidad, y por ende, en el l√≠mite, las proporciones se corresponder√°n con la probabilidad real.
>
> Otra forma de verlo es pensando en un histograma, donde al final sumamos los bins dentro de ese intervalo.

```r
numeric_approx_2 <- function(n, a, b){
    norms <- rnorm(n)
    return (1/n * sum(norms>a & norms<b))
}
```

```r
a <- -1.96
b <-  1.96
numeric_approx_2(a,b,n=100)
numeric_approx_2(a,b,n=1000)
numeric_approx_2(a,b,n=50000)
```

```r
pnorm(q=b)-pnorm(q=a)
```

```r
gridx <- 1:5000
plot(gridx,lapply(gridx, numeric_approx_2, a, b), pch=20, col='orange')
abline(h=pnorm(q=b)-pnorm(q=a), col='red')
```
