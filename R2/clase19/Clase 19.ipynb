```r
datos <- read.table('../clase18/GRB_afterglow.dat.txt', header=T, skip=1)
head(datos)
```

```r
flux <- datos$f
```

$$\large l(\lambda; x) = n log(\lambda) - \lambda \sum x_i$$

```r
EMV_lambda <- function(X){
    l <- 1 / mean(X)
    return (l)
}
```

```r
lambda_est <- EMV_lambda(flux)
lambda_est
```

## 7.



```r
Ft <- function(lambda, t){
    return(1 - exp(-lambda * t))
}
```

```r
t <- 40
```

```r
Ft(lambda_est, t)
```

## 8.

```r
grid_t <- seq(0,130, 0.1)
n <- length(grid_t)
grid_Ft <- rep(NA, n)
for(i in 1:n){
    grid_Ft[i] <- Ft(lambda_est, grid_t[i])
}
```

```r
plot(ecdf(flux), col='steelblue', main="Función de distribución acumulada empírica vs CDF-Exp")
lines(grid_t, grid_Ft, col='orange', lwd=3)
grid()
```
