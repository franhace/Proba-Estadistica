```r
predigo_altura_masculino <- function(alturas, alturasmadres, alturamadrenueva, h){
    
    df <- data.frame(Filas=c(1:length(alturas)), Altura=alturas, AlturaMadre=alturasmadres)
    #print(df[df$Altura >= 155 &  df$Altura <= 190,] )
    #datos_alturas_prediccion <- df[df$Altura >= (alturamadrenueva-h) &
    #                               df$Altura <= (alturamadrenueva+h),]
    datos_alturas_prediccion <- df[df$AlturaMadre >= 155 &
                                   df$AlturaMadre <= 157, 'Altura']
    
    #print(datos_alturas_prediccion)
    prediccion_promedio <- mean(datos_alturas_prediccion)
    return(prediccion_promedio)
}
```

```r
datos_hombres <- data[data$genero == 'M', ]
predigo_altura_masculino(datos_hombres$altura, datos_hombres$altura_madre, 156, 0.5)
```
