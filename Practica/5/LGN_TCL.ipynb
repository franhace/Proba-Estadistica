# Límite de variables aleatorias

### 1. **Ley de los Grandes Números**

```r
N_rep = 1000

sample_means<-rep(NA,N_rep)
for (N in c(1,2,10,50, 1000))
    {
        for(i in 1:N_rep)
        {
            sample<-runif(N,0,1)
            sample_means[i]<-mean(sample)
        }
    hist(sample_means, freq=F, xlim=c(0,1), main=N)
    abline(v=0.5,col='red')
    }
```

### 2. **Teorema central del límite**

```r
N_rep = 1000

sample_means<-rep(NA,N_rep)
for (N in c(1,2,5, 10, 20, 50, 100))
    {
        for(i in 1:N_rep)
        {
            sample<-runif(N,0,1)
            sample_means[i]<-(mean(sample)-0.5)*sqrt(12*N)
        }
    hist(sample_means, freq=F, xlim=c(-4,4), main=N)
    curve(dnorm(x, mean=0, sd=1), add=TRUE, col = 'red')
    #abline(v=0.5,col='red')
    }
```

### 3. **Método de integración Montecarlo**

```r
h = function(x)
    {
        1/(1+x^2)
    }

nsamp = 10000000
mean(h(runif(nsamp, 0, 1)))
```
