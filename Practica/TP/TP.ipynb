```r
# a)
# Generamos la muestra 
n <- 1000
E_unif_0_1 <- 0.5
V_unif_0_1 <- 1/12

set.seed(1)
sample_mean_1 <- runif(n,0,1)

# Histograma
hist(sample_mean_1,probability = TRUE)

# El histograma se parece en densidad a una distribucion uniforme
```

```r
#b)
sample_mean_2 <- replicate(n,mean(runif(2,0,1)))
hist(sample_mean_2,probability = TRUE)
```

```r
# c)
sample_mean_5 <- replicate(n,mean(runif(5,0,1)))
hist(sample_mean_5,probability = TRUE)
```

```r
#d)
sample_mean_30 <- replicate(n,mean(runif(30,0,1)))
hist(sample_mean_30,probability = TRUE)
```

```r
# e) 
sample_mean_500 <- replicate(n,mean(runif(500,0,1)))
hist(sample_mean_500,probability = TRUE)
```

```r
# f) 
sample_mean_1200 <- replicate(n,mean(runif(1200,0,1)))
hist(sample_mean_1200,probability = TRUE)
```

```r
boxplot(sample_mean_1,sample_mean_2,sample_mean_5,sample_mean_30,
        sample_mean_500,sample_mean_1200,
        main = "Promedio de n unif",
        at = c(1,2,3,4,5,6),
        names = c("n = 1",
                  "n = 2",
                  "n = 5",
                  "n = 30",
                  "n = 500",
                  "n = 1200"))
```

```r
# Los promedios se concentran alrededor de 0.5
```

```r
# Normalizamos segun TCL
normalize <- function(x) {
  (x - E_unif_0_1) / sqrt(  (1/12)/n )
}
```

```r
#a normalizado
sample_mean_normalize_1<- sapply(sample_mean_1,FUN=normalize)
hist(sample_mean_normalize_1,probability = TRUE)
boxplot(sample_mean_normalize_1)
```

```r
#b normalizado
sample_mean_normalize_2<- sapply(sample_mean_2,FUN=normalize)
hist(sample_mean_normalize_2,probability = TRUE)
boxplot(sample_mean_normalize_2)
```

```r
#e normalizado
sample_mean_normalize_500<- sapply(sample_mean_500,FUN=normalize)
hist(sample_mean_normalize_500,probability = TRUE)
boxplot(sample_mean_normalize_500)
```

```r
#f normalizado
sample_mean_normalize_1200<- sapply(sample_mean_1200,FUN=normalize)
hist(sample_mean_normalize_1200,probability = TRUE)
boxplot(sample_mean_normalize_1200)
```

```r
boxplot(sample_mean_normalize_1,
        sample_mean_normalize_2,
        sample_mean_normalize_5,
        sample_mean_normalize_30,
        sample_mean_normalize_500,
        sample_mean_normalize_1200,
        main = "Promedio de n unif normalizados",
        at = c(1,2,3,4,5,6),
        names = c("n = 1",
                  "n = 2",
                  "n = 5",
                  "n = 30",
                  "n = 500",
                  "n = 1200"))
```

```r
# h) Cauchy
set.seed(5)
sample_cauchy_1 <- replicate(n,mean(rcauchy(1)))
hist(sample_cauchy_1,probability = TRUE)
sample_cauchy_2 <- replicate(n,mean(rcauchy(2)))
hist(sample_cauchy_2,probability = TRUE)
sample_cauchy_5 <- replicate(n,mean(rcauchy(5)))
hist(sample_cauchy_5,probability = TRUE)
sample_cauchy_30 <-replicate(n,mean(rcauchy(30)))
hist(sample_cauchy_30,probability = TRUE)
sample_cauchy_500 <-replicate(n,mean(rcauchy(500)))
hist(sample_cauchy_500,probability = TRUE)
sample_cauchy_1200 <-replicate(n,mean(rcauchy(1200)))
hist(sample_cauchy_1200,probability = TRUE)
```

```r
c_light <- c("#DCBCBC")
c_light_highlight <- c("#C79999")
c_mid <- c("#B97C7C")
c_mid_highlight <- c("#A25050")
c_dark <- c("#8F2727")
c_dark_highlight <- c("#7C0000")

sorted_sample_cauchy<-sort(sample_cauchy_1)
plot(sorted_sample_cauchy, dcauchy(sorted_sample_cauchy, location = 0, scale = 1), type="l", col=c_dark_highlight, lwd=2,
     main="", xlab="n=1", ylab="Density", yaxt='n')
box()

sorted_sample_cauchy<-sort(sample_cauchy_2)
plot(sorted_sample_cauchy, dcauchy(sorted_sample_cauchy, location = 0, scale = 1), type="l", col=c_dark_highlight, lwd=2,
     main="", xlab="n=2", ylab="Density", yaxt='n')
box()
sorted_sample_cauchy<-sort(sample_cauchy_30)
plot(sorted_sample_cauchy, dcauchy(sorted_sample_cauchy, location = 0, scale = 1), type="l", col=c_dark_highlight, lwd=2,
     main="", xlab="n=30", ylab="Density", yaxt='n')
box()
sorted_sample_cauchy<-sort(sample_cauchy_500)
plot(sorted_sample_cauchy, dcauchy(sorted_sample_cauchy, location = 0, scale = 1), type="l", col=c_dark_highlight, lwd=2,
     main="", xlab="n=500", ylab="Density", yaxt='n')
box()
sorted_sample_cauchy<-sort(sample_cauchy_1200)
plot(sorted_sample_cauchy, dcauchy(sorted_sample_cauchy, location = 0, scale = 1), type="l", col=c_dark_highlight, lwd=2,
     main="", xlab="n=1200", ylab="Density", yaxt='n')
box()
```
