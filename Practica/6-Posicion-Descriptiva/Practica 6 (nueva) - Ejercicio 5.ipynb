```r
ingresos <- read.table("./Datos/ingresos.txt", header = TRUE)
colnames(ingresos)
ingr<-ingresos$total
length(ingr)
```

```r
# a) ¿Cuál es el ingreso mı́nimo percibido por los trabajadores encuestados? 
min(ingr)

# Estime la proporción de los trabajadores de la ciudad que percibe el ingreso mı́nimo.
sum(rle(ingr==15)$values)/length(ingr)
```

```r
# b) Estimar el ingreso mensual que se necesita para pertenecer al 10 % de trabajadores
# de la ciudad con ingresos más altos.
quantile(ingr,0.9)

# el promedio de los que estan en el top 10%
mean(ingr[ingr>=quantile(ingr, 0.9, na.rm=TRUE)], na.rm=TRUE) # tomamos los ingresos que son superior al cuantil .9
```

```r
# c) Calcular la media muestral, la mediana muestral y la media α-podada con α =0,10 (10 %)
mean(ingr)
median(ingr)
mean(ingr,0.1)
```

```r
# d ) Calcular el desvı́o estándar muestral y la distancia intercuartil.
sd(ingr)
IQR(ingr)
```

```r
# e) Realizar un histograma y un boxplot. ¿Cuáles son las caracterı́sticas más sobresalientes? ¿Hay outliers?
hist(ingr)
boxplot(ingr)
```
