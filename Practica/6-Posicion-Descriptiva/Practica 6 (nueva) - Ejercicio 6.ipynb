```r
# a) Generar muestras de tamaño 25, 50 y 100 de una distribución normal. Construir
# QQ–plots para cada una de ellas. Repetir varias veces.
a<-hist(rnorm(25))
b<-hist(rnorm(50))
c<-hist(rnorm(100))
hist(rnorm(1000))
```

```r
data1 <- rnorm(25, mean = 3, sd = 2)
qqnorm(data1)
qqline(data1) 
data2 <- rnorm(50, mean = 3, sd = 2)
qqnorm(data2)
qqline(data2) 
data3 <- rnorm(100, mean = 3, sd = 2)
qqnorm(data3)
qqline(data3) 
data4 <- rnorm(1000, mean = 3, sd = 2)
qqnorm(data4)
qqline(data4) 
```

```r
# b) Repetir a) para una Γ(5, 1/2 ).
data5 <- rgamma(1000, 5, 1/2)
qqnorm(data5)
qqline(data5) 
```

```r
# c) Repetir a) para Y = Z/U
# donde Z ∼ N (0, 1) y U ∼ U(0, 1) independientes.
dataZ <- rnorm(500)
dataU <- runif(500)
y<-dataZ/dataU
qqnorm(y)
qqline(y) 
```

```r
# d ) Repetir a) para una distribución uniforme.

data1 <- runif(25)
qqnorm(data1)
qqline(data1) 
data2 <- runif(50)
qqnorm(data2)
qqline(data2) 
data3 <- runif(100)
qqnorm(data3)
qqline(data3) 
data4 <- runif(1000)
qqnorm(data4)
qqline(data4) 
```

```r
# e) Repetir a) para una distribución exponencial.
data1 <- rexp(25)
qqnorm(data1)
qqline(data1) 
data2 <- rexp(50)
qqnorm(data2)
qqline(data2) 
data3 <- rexp(100)
qqnorm(data3)
qqline(data3) 
data4 <- rexp(1000)
qqnorm(data4)
qqline(data4) 
```
