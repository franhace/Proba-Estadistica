## Estimar el valor de $\pi$

https://seeing-theory.brown.edu/frequentist-inference/index.html

Lanzamos un conjunto de $n$ dardos sobre un tablero circular de radio $r = 1$ metro. El tablero está perfectamente contenido en un cuadrado de lado $l = 2r$. Como tenemos muy mala puntería, los dardos se distribuyen de manera uniforme sobre el tablero y el cuadrado que lo contiene. Notar que la probabilidad $p$ de que un dardo caiga en el tablero es proporcional al área del tablero divido el área total del cuadrado. 

Dado que la distribución de los lanzamientos es uniforme en el cuadrado, sabemos que la probabilidad $p$ de que el lanzamiento caiga en el tablero es

$$p= \frac{A_{tablero}}{A_{cuadrado}}$$

Reemplazando enla fórmula el área de un círculo con radio $r$ y de un cuadrado de lado $l = 2r$, obtenemos que

$p = \frac{\pi}{4}$


Vamos a estimar $\pi$ a partir de $n$ lanzamientos.
Si llamamos $n_{tablero}$ a los dardos que caen en el tablero, 

$\frac{n_{tablero}}{n} \approx \frac{\pi}{4}$

Podemos entonces estimar $\hat \pi = 4\frac{n_{tablero}}{n}$


Generamos las tiradas de los dardos

```r
n = 10000
x <- runif(n)
y <- runif(n)
```

Seleccionamos aquellos que caen dentro del círculo

**Fórmulas útiles**
1. Área del cuadrado: $l^2$
2. Área del círculo: $\pi r^2$
3. Un punto de coordenadas $(x,y)$ está dentro de un círculo de centro $(x_0,y_0)$ y radio $r$ si se cumple la relacion $(x - x_0)^2 + (y - y_0)^2 < r^2$

```r
tablero = sqrt((x - 0.5)**2 + (y - 0.5)**2) < 0.5
```

Contamos cuántos caen adentro del círculo y despejamos para obtener $\pi$

```r
sum(4*tablero)/n
```

Iteramos muchas veces

```r
N_rep = 100000
simulation<-rep(NA, N_rep)
for (n in 1:N_rep)
    {
    x <- runif(n)
    y <- runif(n)
    tablero = sqrt((x - 0.5)**2 + (y - 0.5)**2) < 0.5
    simulation[n] = sum(4*tablero)/n
    }
```

```r

plot(simulation, main="Estimacion de pi",
ylab="pi",
type="l",
col="blue")
abline(h=3.1416,col='red')
```
