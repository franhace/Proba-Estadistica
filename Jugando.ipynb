```r
# Simulo elegir al azar el palo de un mazo de cartas y cuento cuantas de las 100 veces le emboco. 
# Lo hago 1000 veces, porque estamos en cuarentena y no hay nada mas divertido que hacer
juego_mucho<-rbinom(10000000, 100, 0.25)
```

```r
n_100<-head(juego_mucho, 100)
n_1000<-head(juego_mucho, 1000)
n_10000<-head(juego_mucho, 10000)
n_100000<-head(juego_mucho, 100000)
vec<-c(n_100,n_1000,n_10000,n_100000)
mean(n_100)
mean(n_1000)
mean(n_10000)
mean(n_100000)
hist(n_100)
hist(n_1000)
hist(n_10000)
hist(n_100000)
```

```r
# El chaboncito le pego 28 veces al palo
# Quiero saber cual es la proba de hacerlo en cada caso
mean(n_100 >= 28)
mean(n_1000 >= 28)
mean(n_10000 >= 28)
mean(n_100000 >= 28)
```

```r
# Calculamos el IC para la media
n = length(n_1000)
media = sum(n_1000)/n
alfa = 0.05
# Z tiene distrib normal std
Z<-sapply(c(1-alfa/2, alfa/2),
       function(alfa) (2*n*media)/qchisq(alfa ,2*n))
z_alpha_2<-Z[2]
menos_z_alpha_2<-Z[1]
z_alpha_2
menos_z_alpha_2
```

```r
# Bordes IC
hist(n_100000)
abline(v = z_alpha_2, col = "red", lwd = 3)
abline(v = menos_z_alpha_2, col = "red", lwd = 3)
```
