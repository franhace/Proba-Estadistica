```r
datos <- scan("Practica/6/Datos/ingresos.txt")
# datos10<-head(datos,100)
# datos10
summary(datos)
```

```r
#Media 
cat("\nmedia:",mean(datos))
# Media alpha podada 10%
cat("\nmedia alfa podada:",mean(datos,trim=0.1))
#Mediana Muestral
cat("\nmediana muestral:",median(datos))
quantile(datos)
# Varianza
cat("Varianza1:",mean(datos^2)-mean(datos)^2)
cat("\nVarianza2:",mean((datos-mean(datos))^2))
cat("\nUsando var(x):",var(datos))
# Desvio Muestral: raiz cuadrada de la varianza muestral
cat("Desvio muestral:",sqrt(apply(datos, 2, var)))
# Estimacion de la varianza a partir del estimador \hat{sigma}²
cat("\nEstimacion de la varianza a partir del estimador \\hat{sigma}²:",sum((datos-mean(datos))^2)/(n))
# Varianza Muestral S²
n<-1000
cat("\nVar muestral S²:",sum((datos-mean(datos))^2)/(n-1))
cat("\nIQR:",IQR(datos))
cat("\nMAD:",mad(datos, constant=1))
```

```r
# Forma de miniminar EAM
library(L1pack)
with(data=father.son, plot(fheight,sheight))
ajuste2 <- lm(sheight~fheight, data=father.son)
ajuste4 <- l1fit(father.son$fheight,father.son$sheight)
abline(ajuste2$coefficients)
abline(ajuste4$coefficients, col=2)
```
