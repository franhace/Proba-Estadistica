# Cumples
### Cual es la probabilidad de que al menos 2 alumnos (de una clase de n alumnos) cumplan el mismo dia?
##### E = {al menos 2 alumnos cumplan el mismo dia}
##### P(E) = 1 - P(E^c)
##### E^c = {todos los alumnos cumplen en dias distintos}
##### P(E) = 1 - 365*..*(365-(n-1))/365^n

```r
proba_mismo_cumple <- function(n){
    1 - prod(365:(365 - n + 1))/365^n
}

proba_mismo_cumple(10)
```

```r
# miramos que pasa para distintos n's
enes <- seq(2,100,1)

proba_mismo_cumples <- sapply(enes, proba_mismo_cumple)

plot(enes, proba_mismo_cumples, type="l")
```

```r
# Para que numero de alumnos se cumple que la proba es mayor que r
r <- 0.5
x2 <- which(proba_mismo_cumples>r)
x2
```

# Simulaciones

```r
# elige un n elementos de un vector x, con reposicion
# sample(x,n,replace=TRUE)
tiro2dados <- sample(1:6,2,replace=TRUE)
tiro2dados

# me va a dar siempre igual
set.seed(435345)
tiro2dadosfijos <- sample(1:6,2,replace=TRUE)

# replicamos el experimento expr, n veces
# replicate(n, expr)
replicate(5, sample(1:6,2,replace=TRUE))


# lo mismo pero a mano
resultado <- matrix(0, nrow=2, ncol=5)
for(i in 1:5){
    resultado[,i] <- sample(1:6,2,replace=TRUE)
}
resultado
```

### Se tiran 2 dados, cual es la proba de que sumen 8?

```r
sumo2dados <- sum(sample(1:6,2,replace=TRUE))

replicate(100, sum(sample(1:6,2,replace=TRUE)))

# si quiero los valores iguales a 8

replicate(100, sum(sample(1:6,2,replace=TRUE))) == 8

# miro el promedio
mean(replicate(100, sum(sample(1:6,2,replace=TRUE))) == 8)
```
